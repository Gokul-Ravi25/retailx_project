### Load Customers Data to Silver table as Delta

```python
bronze_df = spark.table('retailx.bronze.customers')
bronze_df.printSchema()
bronze_df.show(5)
```

```python
# Load as delta in silver
silver_df = bronze_df.select('customer_id', 'name', 'email', 'created_date')
silver_df.printSchema()
```

```python
# save as silver delta table
silver_df.write.format('delta').mode('overwrite').saveAsTable('retailx.silver.customers')
```
