### Ingest customer data to bronze using Auto Loader

```python

df = (
    spark.readStream
        .format("cloudFiles")
        .option("cloudFiles.format", "csv")
        .option("cloudFiles.schemaLocation", "s3://retailx-datbricks-bucket/checkpoints/bronze/customers")  
        .option("header", "true")
        .option("cloudFiles.inferColumnTypes", "true")
        .load("s3://retailx-datbricks-bucket/bronze/customers/")
)

(
    df.writeStream
      .format("delta")
      .option("checkpointLocation", "s3://retailx-datbricks-bucket/checkpoints/bronze/customers/")
      .option("mergeSchema", "true")
      .trigger(availableNow=True)  
      .toTable("retailx.bronze.customers")
)
```
